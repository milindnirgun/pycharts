### Bin specific column of dataframe to get categorical counts
The following code cuts the "less_than_2_perc" column values into 10 categories. Because this column has percentage values, it conveniently bins it to 0-10%, 10-20%,...,90-100% ranges. Then we can do value_count on this array to get a histogram.

```python
perfranges = pd.cut(perf['less_than_2_perc'], 10)
```

```python
perfranges[:5]
```

```python
perfranges.shape
```

```python
pd.value_counts(perfranges)
```

Repeat the above operation for different types of transactions as the ranges make sense only within transaction categories.

```python
standard = perf[perf['CATEGORY_NAME'] == 'Standard']
s_ranges = pd.cut(standard['less_than_2_perc'], 10)
pd.value_counts(s_ranges)
```

```python
fig, ax = plt.subplots()
a_hist, a_bins = np.histogram(standard['less_than_2_perc'])
b_hist, b_bins = np.histogram(standard['less_than_2.5_perc'], bins=a_bins)
width = (a_bins[1] - a_bins[0])/3
ax.bar(a_bins[:-1], a_hist, width=width, facecolor='cornflowerblue')
ax.bar(b_bins[:-1]+width, b_hist, width=width, facecolor='seagreen')
```

```python
exception = perf[perf['CATEGORY_NAME'] == 'Exception']
e_ranges = pd.cut(standard['less_than_2_perc'], 10)
pd.value_counts(e_ranges)
```

```python
fig, ax = plt.subplots()
a_hist, a_bins = np.histogram(exception['less_than_2_perc'])
b_hist, b_bins = np.histogram(exception['less_than_2.5_perc'], bins=a_bins)
width = (a_bins[1] - a_bins[0])/3
ax.bar(a_bins[:-1], a_hist, width=width, facecolor='cornflowerblue')
ax.bar(b_bins[:-1]+width, b_hist, width=width, facecolor='seagreen')
```
