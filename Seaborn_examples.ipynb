### Exploring Datasets in Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#tips[tips['time'] == "Lunch"]
tips['perc'] = tips.tip/tips.total_bill * 100
tips.head()

#sns.set_style("ticks", {"ytick.minor.size": 2})
f, ax = plt.subplots(figsize=(10, 12))

# Show distribution of percentage of tips for lunch & dinner
sns.violinplot(x = "time", y = "perc", data = tips);
plt.suptitle("Distribution of Tips during Lunch & Dinner", fontsize = 12)
ax.set_ylabel('Percentage of Total Bill')
ax.set_xlabel('Type of Meal')
plt.show()
```

```python
g = sns.FacetGrid(tips, col = "sex", row = "smoker", hue = "time", palette = "Set1",
                 hue_order = ["Dinner", "Lunch"], margin_titles = True)
g = (g.map(plt.scatter, "total_bill", "perc").add_legend())
#g = g.map(plt.violinplot(x = "time", y = "perc")
```
